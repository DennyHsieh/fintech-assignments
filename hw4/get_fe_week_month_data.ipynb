```python
import os
import pandas as pd

csv_dl_folder = 'price_csv_dl'
week_month_folder = 'etf_week_month_data'
if not os.path.exists(week_month_folder):
    os.mkdir(week_month_folder)


def to_df(folder):
    files = os.listdir(folder)
    df_closes = []
    df_empty = True

    for file in files:
        etf_name = file.split('.')[0]
        file = os.path.join(folder, file)

        df = pd.read_csv(file)

        df_close = df[['Date', 'Close']]
        df_close.rename(columns={'Close': etf_name}, inplace=True)

        if df_empty:
            df_closes = df_close
            df_empty = False
            continue

        df_closes = pd.merge(df_closes, df_close, on='Date', how='outer')
    return df_closes


# Financials Equity folder
price_folders = os.listdir(csv_dl_folder)
fe_folder = price_folders[0]
fe_price_folder = os.path.join(csv_dl_folder, fe_folder)

fe_etf_df = to_df(fe_price_folder)
fe_etf_df.head(10)
```

```python
import datetime

per_week_data = []
per_month_data = []


def get_weekday(date_string):
    date_lt = list(map(int, date_string.split('-')))
    date_weekday = datetime.date(date_lt[0], date_lt[1], date_lt[2]).weekday()
    return date_weekday


def get_month(date_string):
    date_lt = list(map(int, date_string.split('-')))
    return date_lt[1]


# Get week and month month with dataframe
for idx, date_string in enumerate(fe_etf_df['Date']):
    if idx == len(fe_etf_df['Date']) - 1:
        continue
    if get_weekday(fe_etf_df['Date'][idx]) > get_weekday(fe_etf_df['Date'][idx + 1]):
        per_week_data.append([fe_etf_df['Date'][idx], idx])
    if get_month(fe_etf_df['Date'][idx]) != get_month(fe_etf_df['Date'][idx + 1]):
        per_month_data.append([fe_etf_df['Date'][idx], idx])
```

```python
print(per_week_data, len(per_week_data))
print(per_month_data, len(per_month_data))
```

```python
week_data_all = {}
month_data_all = {}
for etf in fe_etf_df.columns:
    if etf == 'Date':
        continue
    week_data = []
    month_data = []
    for idx, (_, week_end_idx) in enumerate(per_week_data):
        if idx == 0:
            continue
        this_week = float(fe_etf_df[etf][week_end_idx])
        last_week = float(fe_etf_df[etf][per_week_data[idx - 1][1]])
        week_data.append((this_week - last_week) / this_week)
    for idx, (_, month_end_idx) in enumerate(per_month_data):
        if idx == 0:
            continue
        this_month = float(fe_etf_df[etf][month_end_idx])
        last_month = float(fe_etf_df[etf][per_month_data[idx - 1][1]])
        month_data.append((this_month - last_month) / this_month)

    week_data_all[etf] = week_data
    month_data_all[etf] = month_data

df_week_data = pd.DataFrame(week_data_all)
df_month_data = pd.DataFrame(month_data_all)

df_week_data.to_csv(os.path.join(week_month_folder, 'fe_week_data.csv'), index=False)
df_month_data.to_csv(os.path.join(week_month_folder, 'fe_month_data.csv'), index=False)
```
