```python
import os
import pandas as pd

week_month_folder = 'etf_week_month_data'

fe_week_df = pd.read_csv(os.path.join(week_month_folder, 'fe_week_data.csv'))
fe_week_df.head()
```

```python
fe_month_df = pd.read_csv(os.path.join(week_month_folder, 'fe_month_data.csv'))
fe_month_df.head()
```

```python
cse_week_df = pd.read_csv(os.path.join(week_month_folder, 'cse_week_data.csv'))
cse_week_df.head()
```

```python
cse_month_df = pd.read_csv(os.path.join(week_month_folder, 'cse_month_data.csv'))
cse_month_df.head()
```

## Sharpe Omega measure

```python
# Risk-free rate
week_riskfree = (2.45785 / 100 ) / 52
month_riskfree = (2.45785 / 100 ) / 12

fe_name = fe_week_df.columns
cse_name = cse_week_df.columns
```

```python
def Ep_val_uni(data, rf):
    fil = (rf > data[name])
    return (data[name][fil] - rf).mean()

def Ep_val(data):
    return (data[name].mean())


# Expected values for Financials Equity
fe_expect_val = pd.DataFrame()
for name in fe_name:
    fe_expect_val[name] = [Ep_val(fe_week_df), Ep_val(fe_month_df), Ep_val_uni(fe_week_df, week_riskfree), Ep_val_uni(fe_week_df, month_riskfree)]

fe_EpX = fe_expect_val.T
fe_EpX.columns = ['Expect_X_all_week','Expect_X_all_month','Expect_X_part_week','Expect_X_part_month']
fe_EpX.head()
```

```python
cse_expect_val = pd.DataFrame()
for name in cse_name:
    cse_expect_val[name] = [Ep_val(cse_week_df), Ep_val(cse_month_df), Ep_val_uni(cse_week_df, week_riskfree), Ep_val_uni(cse_week_df, month_riskfree)]

# Expected values for Consumer Staples Equity
cse_EpX = cse_expect_val.T
cse_EpX.columns = ['Expect_X_all_week','Expect_X_all_month','Expect_X_part_week','Expect_X_part_month']
cse_EpX.head()
```

```python
# FE Ranking
fe_rank_weekly = pd.DataFrame() 
fe_rank_monthly = pd.DataFrame()

fe_sharpOmeasure_weekly = pd.DataFrame()
fe_sharpOmeasure_monthly = pd.DataFrame()
fe_sharpOmeasure_weekly["Weekly Sharpe Omega"] = (fe_EpX["Expect_X_all_week"] - week_riskfree) / fe_EpX["Expect_X_part_week"]
fe_sharpOmeasure_monthly["Monthly Sharpe Omega"] = (fe_EpX['Expect_X_all_month'] - month_riskfree) / fe_EpX["Expect_X_part_month"]

fe_rank_weekly["Financials_Equity_Sharpe_Omega_week_rank"] = fe_sharpOmeasure_weekly.sort_values(by = "Weekly Sharpe Omega", ascending= False).index
fe_rank_monthly["Financials_Equity_Sharpe_Omega_month_rank"] = fe_sharpOmeasure_monthly.sort_values(by = "Monthly Sharpe Omega", ascending= False).index
```

### Result of Financials Equity

```python
fe_sharp_concat_df = pd.concat([fe_rank_weekly, fe_rank_monthly], axis=1)
fe_sharp_concat_df.head(15)
```

### Result of Consumer Staples Equity

```python
# CSE Ranking
cse_rank_weekly = pd.DataFrame() 
cse_rank_monthly = pd.DataFrame()

cse_sharpOmeasure_weekly = pd.DataFrame()
cse_sharpOmeasure_monthly = pd.DataFrame()
cse_sharpOmeasure_weekly["Weekly Sharpe Omega"] = (cse_EpX["Expect_X_all_week"] - week_riskfree) / cse_EpX["Expect_X_part_week"]
cse_sharpOmeasure_monthly["Monthly Sharpe Omega"] = (cse_EpX['Expect_X_all_month'] - month_riskfree) / cse_EpX["Expect_X_part_month"]

cse_rank_weekly["Consumer_Staples_Equity_Sharpe_Omega_week_rank"] = cse_sharpOmeasure_weekly.sort_values(by = "Weekly Sharpe Omega", ascending= False).index
cse_rank_monthly["Consumer_Staples_Equity_Sharpe_Omega_month_rank"] = cse_sharpOmeasure_monthly.sort_values(by = "Monthly Sharpe Omega", ascending= False).index
```

```python
cse_sharp_concat_df = pd.concat([cse_rank_weekly, cse_rank_monthly], axis=1)
cse_sharp_concat_df.head(15)
```

```python
rank_folder = 'rank'
if not os.path.exists(rank_folder):
    os.mkdir(rank_folder)

fe_sharp_concat_df.to_csv(os.path.join(rank_folder, 'fe_rank.csv'), index=False)
cse_sharp_concat_df.to_csv(os.path.join(rank_folder, 'cse_rank.csv'), index=False)
```
