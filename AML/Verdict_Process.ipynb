```python
from docx import Document
from docx.shared import Pt
from docx.shared import Inches
from docx.oxml.ns import qn
import docx
from os import listdir
from os.path import isfile, isdir, join
import pandas as pd
```

```python
# 指定要列出所有檔案的目錄
mypath = "./word/txt"

# 取得所有檔案與子目錄名稱
files = listdir(mypath)
files
```

```python
file=docx.Document('C:/Users\Hanwu\Desktop\Govt_web_crawl\Govt_web_crawl\word\最高行政法院 89年度判字第2907號.docx')
print("段落數:" +str(len(file.paragraphs)))
for para in file.paragraphs:
    print(para.text)
#輸出段落編號及段落內容
for i in range(len(file.paragraphs)):
    print("第" +str(i)+ "段的內容是：" +file.paragraphs[i].text)
```

```python
name=[]
crime=[]
news=[]
key_sentence=[]
news_link=[]
content=[]
data=pd.DataFrame(columns=["name","crime","crime_risk","suspect_probability","key_sentence","news","news_title","news_link","other_suspect","content"])
for i in files:
    file_dir =mypath+"/"+i
    f = open(file_dir,encoding = 'utf-8')
    txt_read = f.read()
    txt_read=txt_read.replace('\u3000','')
    txt_read=txt_read.replace('\n','')
    txt_read=txt_read.replace('\ufeff','')
    content.append(txt_read)
    print(i)
    z=i.replace(".txt",'')
    link_behind=z.replace(" ",'%20')
    link ="https://lawsnote.com/search/all/"+link_behind
    str_front = '被告'
    str_behind = '上列'
    nPos_front = max(0,txt_read.index(str_front))
   # nPos_behind = max(0,txt_read.index(str_behind))
    #前後隔出人名
    why_front = '【裁判案由】'
    why_behind = '【裁判內文】'
    whypos_front = txt_read.index(why_front)
    whypos_behind = txt_read.index(why_behind)
    #前後隔出法條
    main_content_front = '主文'
    main_content_behind= '理由'
    main_content_behind＿2='事實'
    main_content_front_pos  = txt_read.index(main_content_front)
    main_content_behind_pos1 = txt_read.index(main_content_behind)
    main_content_behind_pos2 = txt_read.index(main_content_behind＿2)
    main_content_behind_pos=min(main_content_behind_pos1,main_content_behind_pos2)
    #print(nPos_front) 
    #print(nPos_behind) 
    
    name.append(txt_read[nPos_front+len(str_front):nPos_front+len(str_front)+3])#人名
    crime.append(txt_read[whypos_front+len(why_front):whypos_behind])#法條
    news.append(i.replace(".txt",''))#判決書名稱
    news_link.append(link)
    key_sentence.append(txt_read[main_content_front_pos+len(main_content_front):main_content_behind_pos])

    
    
```

```python
data["crime"]=crime
data["name"]=name
data["news"]=news
data["key_sentence"]=key_sentence
data["news_link"]=news_link
data["crime_risk"]="高"
data["content"]=content
data
```

```python
data.to_csv("判決書.csv")
```
